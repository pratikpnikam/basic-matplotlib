```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[23,56,12,55,65,33,23]
```

```python
plt.plot(x,y,color='green',linewidth=2)
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('Weather')
```

```python
plt.plot(x,y,'go')#g i
```

```python
days=[1,2,3,4,5,6,7]
max_t=[55,49,46,43,32,32,44]
min_t=[21,22,23,23,24,25,28]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label='MAX',color='green')
plt.plot(days,min_t,label='MIN',color='blue')
plt.plot(days,avg_t,label='AVG',color='black')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.legend(loc='upper right')  #legend is used to show index in top right corner if there are more than one lines to plot on graph
                                #loc is location
```
